```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.manifold import Isomap
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
iris = load_iris()
X = iris.data
y = iris.target
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
lda = LinearDiscriminantAnalysis(n_components=2,solver='svd')
X_lda = lda.fit_transform(X, y)
```

```python
iso = Isomap(n_components=2)
X_iso = iso.fit_transform(X_scaled)
```

```python
from sklearn.manifold import LocallyLinearEmbedding
lle = LocallyLinearEmbedding(n_components=2)
X_lle = lle.fit_transform(X_scaled)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(13.5, 4))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='Set1', ax=ax[0])
sns.scatterplot(x=X_lda[:, 0], y=X_lda[:, 1], hue=y, palette='Set1', ax=ax[1])
sns.scatterplot(x=X_iso[:, 0], y=X_iso[:, 1], hue=y, palette='Set1', ax=ax[2])
sns.scatterplot(x=X_lle[:, 0], y=X_lle[:, 1], hue=y, palette='Set1', ax=ax[3])
ax[0].set_title('PCA of IRIS', fontsize=15, pad=20)
ax[1].set_title('LDA of IRIS', fontsize=15, pad=20)
ax[2].set_title('Isomap of IRIS', fontsize=15, pad=20)
ax[3].set_title('LLE of IRIS', fontsize=15, pad=20)
```
