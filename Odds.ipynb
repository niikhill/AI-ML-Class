```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#initilizing the Win and Loose List
Win = list(range(1,1000,1))
Lose = list(range(999,0,-1))

#Creating empty data frame
df = pd.DataFrame()

# initializing the columns of the data frames with the two lists
df['Win'] = Win
df['Lose'] = Lose

#Calculating the odds of winning and loosing 
df['Odds_Win'] = df['Win']/df['Lose']
df['Odds_Lose'] = df['Lose']/df['Win']

#Calculating the logs of odds of winning and loosing 

df['Log_Odds_Win'] = np.log(df['Odds_Win'])
df['Log_Odds_lose'] = np.log(df['Odds_Lose'])

#Plotting the logs odds of winning
sns.displot(df['Log_Odds_Win'],kde=True)
plt.title("Distributing of logs of odds in Favour of Winning")
plt.xlabel("Odds Logs")
plt.ylabel("Count")
plt.show()

```
